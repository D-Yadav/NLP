```python
import re
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]',' ',sentences[i]) #Any special caharacter replaced with ' '
    review=review.lower() #changing to lowercase
    corpus.append(review) #and appending to corpus finally
```

```python
## Stemming
for i in corpus:
    words=nltk.word_tokenize(i) #Here first of all we are converting sentences into words
    for word in words:
        if word not in set(stopwords.words('english')): #setting up"stopwords" for word not present on 'english-stopwords'
            print(stemmer.stem(word))
#word_tokenize converts sentence into words
# whichever word is not present in the english "stopwords" we're basically applying stemming for that.
```
